```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/pk664/Downloads/stock_prices.csv')
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df.set_index('date', inplace=True)
```

```python
df
```

```python
rolling_avg = df['closing_price'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['closing_price'], label='Closing Price', color='blue', alpha=0.7)
plt.plot(rolling_avg.index, rolling_avg, label='30-Day Avg', color='red')
```
